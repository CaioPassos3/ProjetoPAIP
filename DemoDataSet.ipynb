```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
file_path = "/content/students_dropout_academic_success.csv"
df = pd.read_csv(file_path)

df.head()
```

Descrição das colunas:

Marital Status: Estado civil do aluno no momento da matrícula.

Application Mode: Método/canal de inscrição do aluno.

Application Order: Ordem de prioridade no envio da inscrição.

Course: Programa de graduação matriculado.

Daytime/Evening Attendance: Cronograma de estudos (diurno/noturno).

Previous Qualification: Qualificação acadêmica antes da matrícula.

Previous Qualification (Grade): Nota/pontuação da educação anterior.

Nationality: País de cidadania.

Mother’s Qualification: Nível educacional da mãe do aluno.

Father’s Qualification: Nível educacional do pai do aluno.

Mother’s Occupation: Tipo/categoria de trabalho da mãe.

Father’s Occupation: Tipo/categoria de trabalho do pai.

Admission Grade: Nota de entrada na universidade.

Displaced: Indica se o aluno foi deslocado (por exemplo, para o serviço militar).

Educational Special Needs: Sinaliza alunos que precisam de suporte de educação especial.

Debtor: Indica se o aluno deve taxas.

Tuition Fees Up to Date: Se as taxas de matrícula são pagas.

Gender: gênero do aluno.

Scholarship Holder: se o aluno recebeu uma bolsa.

Age at Enrollment: idade (em anos) no momento da matrícula.

International: se o aluno é internacional.


Características Acadêmicas (Semestre 1)

Curricular Units 1st Sem (Credited): Disciplinas creditadas no semestre 1.

Curricular Units 1st Sem (Enrolled): Disciplinas matriculadas no semestre 1.

Curricular Units 1st Sem (Evaluations): Provas/avaliações realizadas.

Curricular Units 1st Sem (Approved): Disciplinas aprovadas no semestre 1.

Curricular Units 1st Sem (Grade): Nota média do semestre 1.

Curricular Units 1st Sem (Without Evaluations): Disciplinas sem avaliações.


Características Acadêmicas (Semestre 2)

Curricular Units 2nd Sem (Credited): Disciplinas creditadas no semestre 2.

Curricular Units 2nd Sem (Enrolled): Disciplinas matriculadas no semestre 2.

Curricular Units 2nd Sem (Evaluations): Provas/avaliações realizadas.

Curricular Units 2nd Sem (Approved): Disciplinas aprovadas no semestre 2.

Curricular Units 2nd Sem (Grade): Nota média do semestre 2.

Curricular Units 2nd Sem (Without Evaluations): Disciplinas sem avaliações.


Contexto Econômico

Unemployment Rate: Taxa nacional de desemprego no momento da inscrição.

Inflation Rate: Inflação ao consumidor no momento da inscrição.

GDP: Indicador econômico nacional (ex.: PIB per capita).


Alvo

Target: Situação acadêmica final: Dropout (desistência), Enrolled (inscrito), ou Graduate (graduado).

Distribuição de curso por alvo

```python
# Tabela cruzada entre 'Course' e 'target'
cross_tab_course = pd.crosstab(df['Course'], df['target'])

# Criar gráfico de barras empilhadas
cross_tab_course.plot(kind='bar', stacked=True, figsize=(10, 6))

# Personalização do gráfico
plt.title('Distribuição de Course por Target')
plt.xlabel('Course')
plt.ylabel('Número de Estudantes')
plt.xticks(rotation=90)
plt.tight_layout()
plt.legend(title='Target')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Número de dropouts por curso

```python
# Contagem de dropouts por curso
dropout_por_curso = df[df['target'] == 'Dropout']['Course'].value_counts()

# Gráfico de barras
dropout_por_curso.plot(kind='bar', figsize=(10, 6))
plt.title('Número de Dropouts por Curso')
plt.xlabel('Curso')
plt.ylabel('Quantidade de Dropouts')
plt.xticks(rotation=90)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Porcentagem de desistência por curso

```python
# Filtrar apenas as colunas relevantes
df_filtrado = df[['Course', 'target']]

# Contar total de alunos por curso
total_por_curso = df_filtrado.groupby('Course').size()

# Contar número de desistentes por curso
desistentes_por_curso = df_filtrado[df_filtrado['target'] == 'Dropout'].groupby('Course').size()

# Calcular porcentagem de desistência por curso
percentual_desistencia = (desistentes_por_curso / total_por_curso * 100).fillna(0)

# Organizar do menor para o maior percentual de desistência
percentual_desistencia_sorted = percentual_desistencia.sort_values()

# Plotar gráfico de barras
ax = percentual_desistencia_sorted.plot(kind='bar', figsize=(12, 6))
plt.title('Porcentagem de Desistência por Curso (Ordenado Crescente)')
plt.xlabel('Curso')
plt.ylabel('Porcentagem de Desistência')
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

Distribuição de turno por alvo

Turno 1 - Diurno

Turno 0 - Noturno

```python
# Criar tabela cruzada entre frequência e target
cross_tab_attendance = pd.crosstab(df['Daytime/evening attendance'], df['target'])

# Converter para porcentagens por linha
cross_tab_attendance_pct = cross_tab_attendance.div(cross_tab_attendance.sum(axis=1), axis=0) * 100

# Plotar gráfico de barras empilhadas
ax = cross_tab_attendance_pct.plot(kind='bar', stacked=True, figsize=(8, 5))

# Adicionar valores percentuais sobre as barras
for i, row in enumerate(cross_tab_attendance_pct.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, cumulative + val / 2, f"{val:.1f}%", ha='center', va='center', fontsize=9)
        cumulative += val

# Personalização do gráfico
plt.title('Distribuição Percentual de Target por Tipo de Frequência (Daytime/Evening)')
plt.xlabel('Tipo de Frequência (0 = Noturno, 1 = Diurno)')
plt.ylabel('Porcentagem de Estudantes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.legend(title='Target')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Distribuição da nota de admissão por curso e alvo

```python
# Ajustar tamanho do gráfico
plt.figure(figsize=(14, 6))

# Criar boxplot com seaborn
sns.boxplot(
    data=df,
    x='Course',
    y='Admission grade',
    hue='target'
)

plt.title('Distribuição da Nota de Admissão por Curso e por Target')
plt.xlabel('Curso')
plt.ylabel('Nota de Admissão')
plt.xticks(rotation=90)
plt.legend(title='Target')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
# Recriar a coluna 'Faixa de Nota (%)'
df['Faixa de Nota (%)'] = pd.cut(
    df['Admission grade'],
    bins=[-1, 40, 80, 120, 160, 200],
    labels=['0-20%', '21-40%', '41-60%', '61-80%', '81-100%']
)

# Criar a tabela cruzada
faixas_percentuais = pd.crosstab(df['Faixa de Nota (%)'], df['target'])

# Converter para percentuais por linha
faixas_percentuais_pct = faixas_percentuais.div(faixas_percentuais.sum(axis=1), axis=0) * 100

# Gerar gráfico de barras empilhadas com porcentagens
ax = faixas_percentuais_pct.plot(kind='bar', stacked=True, figsize=(10, 6))

# Adicionar valores percentuais sobre as barras
for i, row in enumerate(faixas_percentuais_pct.values):
    cumulative = 0
    for j, val in enumerate(row):
        if val > 0:
            ax.text(i, cumulative + val / 2, f"{val:.1f}%", ha='center', va='center', fontsize=9)
        cumulative += val

# Personalização do gráfico
plt.title('Distribuição Percentual de Target por Faixas de Nota de Admissão (%)')
plt.xlabel('Faixa de Nota de Admissão (%)')
plt.ylabel('Porcentagem de Estudantes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.legend(title='Target')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Comparação das Notas no 1º Semestre entre Dropouts e Não-Dropouts

```python
# Filtrar os grupos
df_dropout = df[df['target'] == 'Dropout']
df_nao_dropout = df[df['target'] != 'Dropout']

# Criar gráfico lado a lado (subplots)
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Histograma para Dropout
sns.histplot(df_dropout['Curricular units 1st sem (grade)'], bins=30, kde=True, ax=axes[0], color='red')
axes[0].set_title('Notas no 1º Semestre - Dropout')
axes[0].set_xlabel('Nota Média')
axes[0].set_ylabel('Frequência')
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# Histograma para Não-Dropout
sns.histplot(df_nao_dropout['Curricular units 1st sem (grade)'], bins=30, kde=True, ax=axes[1], color='green')
axes[1].set_title('Notas no 1º Semestre - Não-Dropout')
axes[1].set_xlabel('Nota Média')
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

# Ajuste final
plt.suptitle('Comparação das Notas no 1º Semestre entre Dropouts e Não-Dropouts', fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
